```python
URL = 'http://zefix.admin.ch/WebServices/Zefix/Zefix.asmx/SearchFirm'
params = {'name':'Fleury'}
```

```python
req = requests.get(URL, params)
```

```python
begin = req.text.find('<hr />')
end = req.text.find('<hr size="5" />')
```

```python
results = req.text[begin+6:end].split('<p />')
```
